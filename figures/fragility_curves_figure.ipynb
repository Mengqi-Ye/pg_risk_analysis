```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Data
hazard_intensity = [0, 5, 10, 20, 30, 40, 50, 60, 70]
failure_probabilities = [
    [0, 0, 0, 0.04, 0.16, 0.39, 0.61, 0.78, 0.88],
    [0, 0, 0, 0.08, 0.38, 0.64, 0.81, 0.91, 0.96],
    #[0, 0.01, 0.16, 0.5, 0.74, 0.9, 0.96, 0.98],
    [0, 0, 0.02, 0.3, 0.63, 0.82, 0.93, 0.98, 1],
    [0, 0, 0.05, 0.51, 0.82, 0.94, 0.985, 0.993, 1]
]

# Plotting
colors = ['b', 'g', 'm',  'y']
labels = ['Failure 1', 'Failure 2', 'Failure 3', 'Failure 4', 'Failure 5']

for i in range(len(failure_probabilities)):
    x = np.array(hazard_intensity)
    y = np.array(failure_probabilities[i])
    f = interp1d(x, y, kind='cubic')
    x_smooth = np.linspace(min(x), max(x), 500)
    y_smooth = f(x_smooth)
    plt.plot(x_smooth, y_smooth, color=colors[i], label=labels[i])

plt.grid(True)
plt.xticks()
plt.tick_params(axis='x', length=3, which='both', bottom=True, top=False)

plt.xlim(0, 70)
plt.ylim(0, 1)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

plt.tick_params(axis='y', labelsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

plt.figure(figsize=(4.5,2.8))

wind_speeds_50 = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70,75,80])
failure_probs_50 = np.array([0, 0,0,0,0.02, 0.12,0.3,0.49, 0.63,0.73, 0.82, 0.885,0.93,0.96, 0.98, 0.993,1])

f_50 = interp1d(wind_speeds_50, failure_probs_50, kind='cubic')
wind_speeds_50_smooth = np.linspace(min(wind_speeds_50), max(wind_speeds_50), 500)
failure_probs_50_smooth = f_50(wind_speeds_50_smooth)

plt.plot(wind_speeds_50_smooth, failure_probs_50_smooth, color='k',linestyle='-',label='Design wind speed of 50 m/s')

# Scaled wind fragility curve data
scaling_factor = 0.7
wind_speeds_44 = wind_speeds_50 * scaling_factor
failure_probs_44 = failure_probs_50 * scaling_factor

f_44 = interp1d(wind_speeds_44, failure_probs_44, kind='cubic')
wind_speeds_44_smooth = np.linspace(min(wind_speeds_44), max(wind_speeds_44), 500)
failure_probs_44_smooth = f_44(wind_speeds_44_smooth)

plt.plot(wind_speeds_44_smooth, failure_probs_44_smooth, color='k',linestyle=':',label='Design wind speed of 35 m/s')

# Plot both curves on the same graph
#plt.plot(wind_speeds_50, failure_probs_50, label='50 m/s')
#plt.plot(wind_speeds_44, failure_probs_44, label='44 m/s')

# Add labels and legend
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Failure Probability')
plt.legend()

plt.xlim(0, 70)  # x
plt.ylim(0, 1)   # y

plt.xlabel('Wind speed (m/s)',fontsize=14.3)
plt.ylabel('Failure probability',fontsize=14.3)
plt.tick_params(axis='x', labelsize=12.2)
plt.tick_params(axis='y', labelsize=12.2)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Data
p = [0.001, 0.0013, 0.002, 0.004, 0.01, 0.02, 0.04, 0.1, 0.2, 0.4, 0.5, 1]
damage = [0.96, 0.93, 0.88, 0.6, 0.25, 0.17, 0.13, 0.11, 0.1, 0.08, 0.07, 0.02]

f = interp1d(p, damage, kind='cubic')
p_smooth = np.linspace(min(p), max(p), 3000)
damage_smooth = f(p_smooth)

plt.figure(figsize=(4.5,2.8))

plt.plot(np.log10(p_smooth), damage_smooth, color='orange', linestyle='-', linewidth=2)
plt.fill_between(np.log10(p_smooth), damage_smooth, color='orange', alpha=0.3)

highlighted_points = [0.002, 0.01, 0.02, 0.1]

for point in highlighted_points:
    index = p.index(point)
    plt.plot([np.log10(p[index]), np.log10(p[index])], [0, damage[index]], color='black', linestyle='--', linewidth=1.5, alpha=0.5)
    plt.plot([np.log10(p[index]), np.log10(p_smooth[0])], [damage[index], damage[index]], color='black', linestyle='--', linewidth=1.5, alpha=0.5)
    plt.text(-3.05, damage[index], f'D{highlighted_points.index(point)+1}', color='black', fontsize=10, ha='right', va='center')
    plt.text(np.log10(p[index]), -0.02, f'{point}', color='black', fontsize=10, ha='center', va='top')

plt.grid(True, color='lightgray', alpha=0.3)

plt.xlim(-3, 0)  # Set the range for the x-axis (log scale)
plt.ylim(0, 1)   # Set the range for the y-axis

#plt.ylabel('Direct damage', fontsize=12)

plt.xticks(np.log10(p), p)  # Set the x-axis tick labels to the corresponding values in p
plt.tick_params(axis='y', labelleft=True, labelsize=10)  # Show y-axis tick labels on the left side

plt.text(0, -0.1, 'Exceedance probability', color='black', fontsize=12, ha='right', va='top')
plt.text(-3.6, 1.07, 'Direct\ndamage', color='black', fontsize=12, ha='left', va='top')

bbox_props = dict(boxstyle='square,pad=0.3', facecolor='white', edgecolor='black')
plt.text(-2.45, 0.33, 'Risk', color='black', fontsize=12, ha='center', va='center', bbox=bbox_props)

plt.xticks([])
plt.yticks([])

plt.show()

```
